```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot  (diagrams de dispersion)
from turtle import title


x = np.arange(1, 20)
y = x**2
y1 = x**2 + 5*x

#plt.scatter(x, y)
#plt.scatter(x, y, color='red')
#plt.scatter(x, y, color='red', marker='*')
#plt.scatter(x, y, color='red', marker='s')
plt.scatter(x, y, color='red', marker='s', label='x^2', s=10)
plt.scatter(x, y1, color='blue', marker='o', label='x^2+5x', s=10)
plt.legend(title='funciones')
plt.title('Tipo de funciones', fontsize=20)
plt.xlabel('x', color='green')
plt.ylabel('y')
plt.grid()
#plt.xlim(0, 5)

plt.text(5, 300,'Este texto', color= 'brown')

# Agregar las coordenadas de los marcadores en tecto
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x}, {val_x**2})')
```

```python
# histogramas
media = 20
std = 3
numeros = np.random.normal(media, std, 1000)
plt.hist(numeros, bins=20, cumulative=False)
plt.title('Distribución normal')
plt.xlabel('Números')
plt.ylabel('Frecuencia')
```

```python
#plt.boxplot(numeros)
plt.boxplot(numeros, vert=False)
```
